# Feature Extraction and Engineering  

```python
features_to_keep = ['year', 'cement_co2', 'co2', 'coal_co2', 'flaring_co2', 'gas_co2', 'oil_co2']  # Consumption_co2, 

df = df[features_to_keep]
```

**Reasoning for the chosen features** 

Since we are tryng to predict co2 emissions based on different industries, we can already exclude the population and GDP, because this factors have nothing to do with the industry-related emissions and will have a negative effect on accurately predicting the emissions trough industries. 

For this notebook we will only take features directly related to an Industry, also of every industry we will only take 1 suitable feature to start with. 

For example, 'co2_growth_prct' = Annual percentage growth in total production-based emissions of carbon dioxide (CO₂), since this is the total amount of all production-based emissions together, this will not give us a clearly overview of which of the productions contributed the most to the total amount of co2 emissions. These features are not sector related but rather give us an overall overview, which will not help us. 

Also a reason to not use this the amounts is because our target variable which we are trying to predict will be co2, which is already measured as the amount of total production-based emissions of co2 

Once again, here you can find the explenations: https://github.com/owid/co2-data/blob/master/owid-co2-codebook.csv

```python
# Chekking for missing values 

null_values = df.isnull().sum()

nan_values = df.isna().sum()

print("Null-values:")
print(null_values)

print("\nNaN-values:")
print(nan_values)
```

```python
# Replacing our missing values with the mean of that feature 

mean_values = df.mean()

for feature in df.columns:
    df[feature].fillna(mean_values[feature], inplace=True)

print(df)
```

```python
# Chekking for missing values to see if we correctl replced them 

null_values = df.isnull().sum()

nan_values = df.isna().sum()

print("Null-values:")
print(null_values)

print("\nNaN-values:")
print(nan_values)
```
